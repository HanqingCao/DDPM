```python
import torch
from torchvision import datasets, transforms
from denoising_diffusion_pytorch import Unet, GaussianDiffusion, Trainer

# 设置 Google Drive 的挂载路径
google_drive_path = '/content/gdrive/MyDrive/'

# MNIST 数据集加载
transform = transforms.Compose([transforms.ToTensor()])
mnist_dataset = datasets.MNIST(root=google_drive_path, train=True, download=True, transform=transform)
dataloader = torch.utils.data.DataLoader(mnist_dataset, batch_size=12, shuffle=True)

# U-Net 模型
model = Unet(
    dim=64,
    dim_mults=(1, 2, 4, 8),
    flash_attn=True
)

# Gaussian diffusion 使用 U-Net 作为生成器
diffusion = GaussianDiffusion(
    model,
    image_size=32,
    timesteps=1000,
    sampling_timesteps=250
)

# 训练器设置
trainer = Trainer(
    diffusion,
    google_drive_path,
    train_batch_size=12,
    train_lr=8e-5,
    train_num_steps=700000,
    gradient_accumulate_every=2,
    ema_decay=0.995,
    amp=True,
    calculate_fid=True
)

# 训练循环
if __name__ == '__main__':
    trainer.train()
```
